```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
prompts = [
    {"role":"user", "content":"Hello, My name is Prateek Mishra."},
    {"role":"ai", "content":'Hello, Prateek Mishra! How can I assist you today?'},
    {"role":"user", "content":"What is My name ?"}
]

res = llm.invoke(prompts)
```

### Managing History

```python
history = [] ## [{"role":"user", "content":"Who is pm of india ?"}]
while True:
    query = input("User: ")
    if query.lower() in ["exit", "quit", "bye"]:
        print("Good Bye 👋")
        break

    history.append({"role":"user", "content":query})
    print("User: ", query)


    res = llm.invoke(history)
    history.append({"role":"ai", "content":res.content})
    print("AI: ", res.content, "\n")
```
