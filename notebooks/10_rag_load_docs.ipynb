### text file loader

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("../data/speech.txt")
texts = loader.load()

print(texts[0].page_content[:100])
```

### PDF docs loader

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("../data/medical_report.pdf")
docs = loader.load()
print(docs[0].page_content[:100])
```

### CSV FILE LOADER

```python
from langchain_community.document_loaders import CSVLoader
loader = CSVLoader("../data/employee_records.csv")
docs = loader.load()
print(docs[0].page_content[:100])
```

### web Page Loader


```python
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader(web_path="https://techsimplus.com/programs/1-Year-Placement-Oriented-Full-Stack-Developer-Program")
docs = loader.load()
print(docs[0].page_content)
```

### wekipedia loader

```python
from langchain_community.document_loaders import WikipediaLoader
loader = WikipediaLoader(query="GEN AI", load_max_docs=2)
docs = loader.load()
print(docs[0].page_content)
```
