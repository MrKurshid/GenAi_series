```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a:int,b:int):
    """
        it will return sum of two numbers

        Args:
            a: Number one
            b: Number two

        Returns:
            int: The sum of a and b
    """
    return a + b

@tool
def multiply_numbers(a:int,b:int):
    """
        it will return product of two numbers

        Args:
            a: Number one
            b: Number two

        Returns:
            int: The product of a and b
    """
    return a * b
```

```python
add_numbers.invoke({"a":100,"b":11})
```

```python
from langchain_groq import ChatGroq
from langchain.agents import create_agent
```

```python
llm = ChatGroq(model="openai/gpt-oss-20b")
```

```python
agent = create_agent(
    model=llm, 
    tools=[add_numbers, multiply_numbers],
    system_prompt="You are a helpful assistant that can perform basic arithmetic operations, always use the available tools.")
```

```python
response = agent.invoke({"messages":[{"role":"user","content":"What is 2+10*2?"}]})
print(response['messages'][-1].content)
```
