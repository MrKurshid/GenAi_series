```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text = "In the realm of artificial intelligence, the concept of knowledge is multifaceted and complex. Knowledge can be defined as the information, understanding, and skills that an AI system possesses, which enables it to perform tasks, make decisions, and solve problems. It encompasses both explicit knowledge, which is codified and easily articulated, and tacit knowledge, which is more intuitive and harder to express."
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=50,
    chunk_overlap=20
)
```

```python
chunks = splitter.split_text(text)
chunks
```

### document splitter

```python
from langchain_community.document_loaders import TextLoader, PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter( chunk_size = 500, chunk_overlap = 20 )
```

```python
textloader = TextLoader("..//data/speech.txt")
texts = textloader.load()
```

```python
splittedText = splitter.split_documents(texts)
len(splittedText)
splittedText
```

```python
pdfloader = PyPDFLoader("..//data/medical_report.pdf")
pdfdata = pdfloader.load()
```

```python
splittedpdf = splitter.split_documents(pdfdata)
len(splittedpdf)
```
